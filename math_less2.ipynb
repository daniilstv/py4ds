1. Задание 
Напишите уравнение параболы, проходящей через три точки (x,y):
(1,2), (3,10), (5,1)


```python
'''
Уравнение парабалы вида y = ax2 + bx + c

Подставляем координаты:
2 = a*1**2 + b*1 + c
10 = a*3**2 + b*3 + c
1 = a*5**2 + 5*b + c

Приводим к матричному виду:
a + b + c - 2 = 0
9*a + 3*b + c - 10 = 0
25*a + 5*b + c - 1 = 0
''' 
```

```python
# Решая систему уравнений через матрицу получим коэффициенты уравнений матрицы  
z = solve([a + b + c - 2, 9*a + 3*b + c - 10, 25*a + 5*b + c - 1], [a,b,c])
z
```

```python
# create 1000 equally spaced points between -10 and 10
x = np.linspace(-100, 100, 1000)

# уравнение параболы
y = z[a]*x**2 + z[b]*x + z[c]  

# график параболы
plt.axis([-50, 50, -2500, 100])
plt.plot(x, y)
```
