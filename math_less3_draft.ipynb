```python
# http://inp.nsk.su/~grozin/python/python6.html
# https://stackoverflow.com/questions/7965743/how-can-i-set-the-aspect-ratio-in-matplotlib
# https://toster.ru/q/220495
```

```python
# 3.1 график окружности
# (x – x0)**2 + (y – y0)**2 = r2
# (х - 0)**2 + (у - 0)**2 = 16
# x**2 + y**2 = 16

x0 = 0
y0 = 0
r = 4

x = []
y = []

rng = np.linspace(-r, r, 101)

for i in range(len(rng)):
    x_tmp = rng[i]
    x.append(x_tmp)
    y_tmp = math.sqrt(r**2 - x_tmp**2)
    y.append(y_tmp)

for i in range(len(rng)):
    x_tmp = rng[~i]
    x.append(x_tmp)
    y_tmp = -1*math.sqrt(r**2 - x_tmp**2)
    y.append(y_tmp)
    
plt.plot(x, y)
plt.axes().set_aspect(1)
plt.show()    
```

```python
# подсмотренный пример
t = np.arange(0,2*np.pi,0.01)
r=4
plt.plot(r*np.sin(t),r*np.cos(t),lw=3)
plt.axis('equal')
plt.show()
```

```python
# 3.2 график эллипса
# x**2/a**2 + y**2/b**2 = 1 , a > b
# c = sqrt(a**2-b**2)

a = 16
b = 9
c = math.sqrt(a**2-b**2)

x = []
y = []

rng_x = np.linspace(-1*math.sqrt(a), math.sqrt(a), 101)

for i in range(len(rng_x)):
    x_tmp = rng_x[i]
    x.append(x_tmp)
    y_tmp = math.sqrt(b/a*(a-x_tmp**2)) 
    y.append(y_tmp)

for i in range(len(rng_x)):
    x_tmp = rng_x[~i]
    x.append(x_tmp)
    y_tmp = -1*math.sqrt(b/a*(a-x_tmp**2)) 
    y.append(y_tmp)
    
plt.plot(x, y)
plt.axes().set_aspect(1)
plt.show()    

```

```python
def drow_plot(f, n=100, a = 0):
    
    xmin = -1 * n
    xmax = n + 1
    dx = 1
    x = np.around(np.arange(xmin, xmax, dx) , decimals=4)

    y = eval(f)
    
    if a == 0:
        aspect = 'auto'
    else:
        aspect = 1

    plt.plot(x, y)
    plt.axes().set_aspect(aspect)
    plt.show()
```

```python
# тест функции
drow_plot("math.pi*(x**2)",20)
```
